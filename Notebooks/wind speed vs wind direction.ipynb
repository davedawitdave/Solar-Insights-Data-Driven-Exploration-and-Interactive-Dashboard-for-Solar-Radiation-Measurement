```python
chart = Charts(data)
chart.polar_plot(speed_col='WS', direction_col='WD')
```

```python
# Histogram for Wind Speed
chart.histogram(column='WS', bins=20, title="Wind Speed Distribution", xlabel="Wind Speed (m/s)")

# Histogram for Wind Direction
chart.histogram(column='WD', bins=20, title="Wind Direction Distribution", xlabel="Wind Direction (Degrees)")
```

```python
# List of columns for which to create histograms
columns = [
    'GHI', 
    'DNI', 
    'DHI', 
    'ModA',
    'ModB',
    'RH',
    'Precipitation',
    'Tamb',   
    'TModA',  
    'TModB'
]

titles = ["GHI", "DNI", "DHI", "ModA (W/m)", "ModA (W/m)", "Relative Humidity(%)", "Precipitation", "Ambient Temp", "Module Temp A", "Module Temp B"]
labels = columns
chart.histogram(column=columns, bins=8, title=titles, xlabel=labels,figsize=(12, 14), num_cols=2)
```
