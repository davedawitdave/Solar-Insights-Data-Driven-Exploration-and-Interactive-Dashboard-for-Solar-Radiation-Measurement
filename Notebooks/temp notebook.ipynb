```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
data = sns.load_dataset("tips")

# Advanced box plot with hue, palette, and custom styling
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=data, 
            palette="coolwarm", showfliers=False,
            boxprops={'facecolor':'None'})  # transparent boxes
sns.swarmplot(x="day", y="total_bill", hue="smoker", data=data, 
              dodge=True, color=".25", alpha=0.6)
plt.title("Advanced Box Plot with Seaborn")
plt.show()
```
